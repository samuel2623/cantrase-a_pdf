```python
import os
from pikepdf import Pdf, Encryption, Permissions

password = "12345"
folder_path = "/content/drive/MyDrive/Curso_automatizacion/2_gestion_archivos_carpetas/estudiantes_semillleros/certificados"

print("=== PROTEGIENDO PDFs CON CONTRASE√ëA ===\n")

# Configurar permisos restrictivos
permissions = Permissions(
    accessibility=True,        # Permite herramientas de accesibilidad
    extract=False,            # NO permite extraer texto/im√°genes
    modify_annotation=False,   # NO permite modificar anotaciones
    modify_assembly=False,     # NO permite modificar estructura
    modify_form=False,         # NO permite modificar formularios
    modify_other=False,        # NO permite otras modificaciones
    print_lowres=False,        # NO permite impresi√≥n baja calidad
    print_highres=False        # NO permite impresi√≥n alta calidad
)

# Verificar que la carpeta existe
if not os.path.isdir(folder_path):
    print(f"‚ùå Error: Carpeta no encontrada en {folder_path}")
else:
    success_count = 0
    error_count = 0

    # Procesar cada archivo PDF
    for filename in os.listdir(folder_path):
        if filename.endswith(".pdf") and not filename.endswith("_PROTECTED.pdf"):
            filepath = os.path.join(folder_path, filename)
            output_filepath = filepath.replace(".pdf", "_PROTECTED.pdf")

            try:
                # Abrir PDF original
                pdf = Pdf.open(filepath)

                # Crear encriptaci√≥n con sintaxis correcta
                encryption = Encryption(
                    owner=password,      # Contrase√±a del propietario
                    user=password,       # Contrase√±a del usuario
                    R=4,                # AES 128-bit encryption
                    allow=permissions,   # Permisos restrictivos
                    aes=True,           # Usar AES encryption
                    metadata=True       # Encriptar metadatos
                )

                # Guardar PDF protegido
                pdf.save(output_filepath, encryption=encryption)
                print(f"‚úÖ PROTEGIDO: {filename}")
                success_count += 1

            except Exception as e:
                print(f"‚ùå ERROR con {filename}: {e}")
                error_count += 1

    print(f"\n=== RESUMEN ===")
    print(f"‚úÖ Archivos protegidos exitosamente: {success_count}")
    print(f"‚ùå Errores: {error_count}")
    print(f"\nüìÅ Los archivos protegidos terminan en '_PROTECTED.pdf'")
    print(f"üîê Contrase√±a requerida: {password}")

    print(f"\n‚ö†Ô∏è  IMPORTANTE:")
    print(f"   ‚Ä¢ Google Drive puede NO pedir contrase√±a")
    print(f"   ‚Ä¢ Descarga un archivo y √°brelo en tu computadora")
    print(f"   ‚Ä¢ Usa Adobe Reader, Foxit, o visor PDF local")
    print(f"   ‚Ä¢ Los archivos est√°n configurados para NO permitir:")
    print(f"     - Impresi√≥n")
    print(f"     - Copia de texto")
    print(f"     - Modificaciones")

print(f"\n=== VERIFICACI√ìN R√ÅPIDA ===")
# Verificar algunos archivos protegidos
protected_files = [f for f in os.listdir(folder_path) if f.endswith(".pdf")]

if protected_files:
    print(f"üìã Archivos _PROTECTED.pdf encontrados: {len(protected_files)}")

    # Probar el primero
    test_file = os.path.join(folder_path, protected_files[0])
    try:
        # Intentar abrir sin contrase√±a (deber√≠a fallar)
        pdf = Pdf.open(test_file)
        print(f"‚ö†Ô∏è  ADVERTENCIA: {protected_files[0]} se abri√≥ sin contrase√±a")
    except:
        print(f"‚úÖ CONFIRMADO: {protected_files[0]} requiere contrase√±a")

        # Probar con contrase√±a
        try:
            pdf = Pdf.open(test_file, password=password)
            print(f"‚úÖ CONFIRMADO: Se puede abrir con contrase√±a '{password}'")
        except Exception as e:
            print(f"‚ùå ERROR abriendo con contrase√±a: {e}")
else:
    print("‚ùå No se encontraron archivos _PROTECTED.pdf")
```
